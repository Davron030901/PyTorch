```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torchvision.transforms as T
from PIL import Image
```

```python
img=Image.open("/content/dog.jpg")
plt.imshow(img)
```

```python
print(img.size)
```

```python
resize=T.Resize((200,300))
resize_img=resize(img)
print(resize_img.size)
```

Image chegara ramka qo'yib beradi

```python
pad=T.Pad(20)
pad_img=pad(img)
print(pad_img.size)
```

```python
plt.imshow(pad_img)
```

```python
ccrop=T.CenterCrop(300)
center_img=ccrop(img)
print(center_img.size)
plt.imshow(center_img)
plt.show()
```

Image bo'laklaydi

```python
fcrop=T.FiveCrop(100)
f_img=fcrop(img)
print(type(f_img))
print(  f_img[0].size)
print(len(f_img))
```

```python
plt.imshow(f_img[4])
```

```python
plt.figsize=(20,30)
fig,ax=plt.subplots(3,2)
for i in range(5):
  ax[i//2,i%2].imshow(f_img[i])
  if i==0:
    ax[i//2,i%2].set_title("Top left")
  if i==1:
    ax[i//2,i%2].set_title("Top right")
  if i==2:
    ax[i//2,i%2].set_title("Bottom left")
  if i==2:
    ax[i//2,i%2].set_title("Bottom right")
  if i==4:
    ax[i//2,i%2].set_title("Center")
```

```python
gray=T.Grayscale(num_output_channels=1)
gray_img=gray(img)
print(gray_img.size)
```

```python
plt.imshow(gray_img)
```

```python
plt.imshow(gray_img,cmap="gray")
```

```python
fig,ax=plt.subplots(1,5)
for i in range(5):
  ax[i%5].imshow(f_img[i])
  if i==0:
    ax[i%5].set_title("Top left")
  if i==1:
    ax[i%5].set_title("Top right")
  if i==2:
    ax[i%5].set_title("Bottom left")
  if i==3:
    ax[i%5].set_title("Bottom right")
  if i==4:
    ax[i%5].set_title("Center")
```

Rasmni kerakli shaklga olib kelish transform compose

```python
resize=T.Resize(200)
ccrop=T.CenterCrop(150)
transform=T.Compose([ccrop,resize])
transform_img=transform(img)
print(transform_img.size)
```

```python
plt.imshow(transform_img)
```
