```python
arr1=torch.arange(10)
print(arr1)
arr2=torch.arange(1,6)
print(arr2)
arr3=torch.arange(1,10,2)
print(arr3)
arr4=torch.arange(1,10.5,0.5)
print(arr4)
```

```python
arr_zero1=torch.zeros(2)
print(arr_zero1)
arr_zero2=torch.zeros(2,3)
print(arr_zero2)
arr_zero3=torch.zeros(2,3,4)
print(arr_zero3)
```

```python
arr_ones1=torch.ones(2)
print(arr_ones1)
arr_ones2=torch.ones(2,3)
print(arr_ones2)
arr_ones3=torch.ones(2,3,4)
print(arr_ones3)
```

```python
arr_full=torch.full(size=(5,5),fill_value=42)
print(arr_full)
```

```python
arr_linspace=torch.linspace(start=0,end=10,steps=7)#(0,10,7)
print(arr_linspace)
```

```python
arr=[
    [1,2,3,3,56,76,674],
    [4,5,6,54,5,54,45],
    [7,8,9,34,64,98,68]
]
arr=torch.tensor(arr)
print(arr[:,1])
print(arr[1,:])
print(arr[2,2:5])
print(arr[::2,6])
```

```python
data=torch.tensor([3,4,2,221,45,6])
x=torch.tensor([1,2,3,4,5,6])
print(data+x)
print(data*x)
print(data/x)
print(data-x)
print(data//x)
print(data%x)
```

```python
y=torch.tensor([2])
print(data+y)
print(data-y)
print(data*y)
print(data/y)
print(data//y)
print(data%y)
```

```python
z=torch.tensor([
    [43,65,6,67],
    [67,78,78,9]
])
print(y+z)
print(y*z)
print(y/z)
print(y-z)
print(y//z)
print(y%z)
```

```python
data_1=torch.tensor([5,6,7,8,9,10,1,2,3,4],dtype=torch.float)
print(data_1.sum())
print(data_1.prod())# ko'paytmasi
print(data_1.max())
print(data_1.min())
print(data_1.argmax())# max indexsi
print(data_1.argmin())#min indexsi
print(data_1.sort())
print(data_1.median())
print(data_1.unique())
print(data_1.mode())
print(data_1.abs())
print(data_1.mean())
print(data_1.std())
```
